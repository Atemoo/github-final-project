```python
scipy.stats.ttest_ind(boston_df['MEDV'],
                  boston_df['CHAS'], equal_var = True)
```

**Conclusion:** Since the p-value is greater than alpha value 0.05, we fail to reject the null hypothesis as there is no proof that there is a statistical difference in teaching evaluations based on gender


```python
scipy.stats.levene(boston_df['MEDV'],
                   boston_df['AGE'], 
                             center='mean')
```

**Conclusion:** Since the p-value is greater than 0.05, we will fail to reject the null hypothesis as there is no significant evidence that at least one of the means differ.


```python
scipy.stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
```

**Conclusion:** Since the p-value  (Sig. (2-tailed)  > 0.05, we fail to reject  the Null hypothesis and conclude that there  no relationship between  Nitric oxide concentrations and proportion of non-retail business acres per town


```python
import statsmodels.api as sm
```

```python
## X is the input variables (or independent variables)
X = boston_df['DIS']
## y is the target/dependent variable
y = boston_df['MEDV']
## add an intercept (beta_0) to our model
X = sm.add_constant(X) 

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

# Print out the statistics
model.summary()
```

**Conclusion:** Like the t-test, the p-value is less than the alpha (Î±) level = 0.05, so we reject the null hypothesis as there is evidence that there is a difference in the five Boston employment centres on the median value of owner occupied homes.
